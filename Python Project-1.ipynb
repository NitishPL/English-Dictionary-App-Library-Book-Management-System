# Library Management System

```python
from tkinter import *
import sqlite3
import tkinter.ttk as  ttk
import tkinter.messagebox as tkMessageBox
from tkinter import messagebox
```

```python
class Database:
    def __init__(self,db):
        self.con=sqlite3.connect(db)
        self.cur=self.con.cursor()
        sq='CREATE TABLE IF NOT EXISTS books(book_title TEXT,author TEXT,year_of_publication INTEGER,isbn INTEGER)'
        self.cur.execute(sq)
        self.con.commit()
    def insert(self,title,author,year,isbn):
        self.cur.execute('insert into books values(?,?,?,?)',(title,author,year,isbn))
        self.con.commit()
    def fetch(self):
        self.cur.execute('SELECT * from books')
        rows=self.cur.fetchall()
        return rows
    def delete(self,title):
        self.cur.execute('delete from books where book_title=?',(title,))
        self.con.commit()
    def update(self,title,author,year,isbn):
        self.cur.execute('UPDATE books SET author=?,year_of_publication=?,isbn=? where book_title=?',(title,author,year,isbn))
        self.con.commit()
    def exit():
        result =tkMessageBox.askquestion('Do you want to exit ?(y/n)',icon='warning')
        if result == 'yes':
            root.destroy()
            exit()
o=Database('Library_book_management.db')
```
